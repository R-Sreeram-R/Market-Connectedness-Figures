```python
corr_subset_numpy = corr_subset.to_numpy()
corr_subset_numpy
```

```python
import networkx as nx
```

```python
G = nx.Graph()
rows, cols = len(corr_subset_numpy), len(corr_subset_numpy[0])

for i in range(rows):
    G.add_node(i)
```

```python
for i in range(rows):
    for j in range(cols):
        if abs(corr_subset_numpy[i][j]) > 0.5:
            G.add_edge(i, j)
```

```python
pos = nx.circular_layout(G)  # You can adjust the value of k to control the space between nodes
plt.figure(figsize=(8, 8))
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=300, node_color=range(34), font_color='black', font_size=8, edge_color='gray', width=1, arrowsize=10,arrows=True, arrowstyle='-|>')
plt.title("Spllover Matrix ")
plt.axis("off")  # Remove axes and labels
plt.show()
```

```python
def diebold_yilmaz_connectedness(spillover_matrix):
    
    n = len(spillover_matrix)
    dy_connectedness_matrix = np.zeros((n, n))

    for i in range(n):
        for j in range(n):
            if i != j:
                dy_connectedness_matrix[i, j] = np.sqrt(np.sum((spillover_matrix[i] - spillover_matrix[j]) ** 2) / len(spillover_matrix))

    return dy_connectedness_matrix
```

```python
connectedness_matrix = diebold_yilmaz_connectedness(corr_subset_numpy)
```

```python
covariance_matrix = np.cov(returns.T)
spillover_matrix = covariance_matrix / np.diag(covariance_matrix)
spillover_matrix
```

```python
G = nx.Graph()
rows, cols = spillover_matrix.shape

for i in range(rows):
    G.add_node(i)

rows, cols
```

```python
for i in range(rows):
    for j in range(cols):
        if abs(spillover_matrix[i][j]) >= 1:
            G.add_edge(i, j)
```

```python
pos = nx.circular_layout(G)  # You can adjust the value of k to control the space between nodes
plt.figure(figsize=(8, 8))
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=300, node_color=range(34), font_color='black', font_size=8, edge_color='gray', width=1, arrowsize=10,arrows=True, arrowstyle='-|>')
plt.title("Sipllover Matrix ")
plt.axis("off")  # Remove axes and labels
plt.show()
```
